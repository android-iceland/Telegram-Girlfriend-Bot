```python
# @title ##**Required functions** { display-mode: "form" }

import os
import telebot
import numpy as np
API_KEY = '12345675:fadfsabufhgaiugdhiuahiuhgiuahruithiuar'  #@param {type: "string"}


bot = telebot.TeleBot(API_KEY)

db={"output_language": "en","database":False,"old_data":"","girl_name":"Billie Eilish","gf_mode":"off","mode":"chat"}
Your_crush_name = 'Billie Eilish'  #@param {type: "string"}
db['girl_name']=Your_crush_name
def database_check(msg):
  database_commad=False
  if msg=="enable database":
    db["database"]=True
    database_commad=True
  if msg=="disable database":
    db["database"]=False
    database_commad=True
  return database_commad

def output_language_set(out_lang):
  custom_commad=False
  if out_lang == "language english":
      db["output_language"] = "en"
      custom_commad=True
  elif out_lang == "language hindi":
      db["output_language"] = "hi"
      custom_commad=True
  elif out_lang == "language bengali":
      db["output_language"] = "bn"  
      custom_commad=True 
  elif out_lang == "language chinese":
      db["output_language"] = "zh-cn"  
      custom_commad=True 
  elif out_lang == "language japanese":
      db["output_language"] = "ja"  
      custom_commad=True 
  # text=language 
  return custom_commad

from googletrans import Translator
translator = Translator()
def translate(msg):
  trans_msg=translator.translate(msg, dest=db["output_language"])
  trans_text=str(trans_msg.text)
  return trans_text  

def convert_english(message):
  if db['output_language']!="en":
    trans_msg=translator.translate(message, dest="en")
    trans_text=str(trans_msg.text)
    give_text=trans_text
  else:
    give_text= message
  return give_text

from datetime import datetime
import pytz
def get_time():
  UTC = pytz.utc
  IST = pytz.timezone('Asia/Kolkata')
  datetime_ist = datetime.now(IST)
  current_time=str(datetime_ist.strftime("%I:%M %p"))
  date=str(datetime_ist.strftime('%d:%m:%Y'))
  return current_time,date  


import emoji
import demoji
demoji.download_codes()
clear_output()
def extract_emojis(s):
  new_text=""
  for c in s:
    
    if c in emoji.UNICODE_EMOJI['en']:
      try:
        extract=demoji.findall_list(c)
        # print(extract[0])
        new_text=new_text+extract[0]+""
      except:
        pass
    else:
      new_text=new_text+c+""
    
  return new_text

emoji_dict={"joy":["ðŸ˜‚"],"pensive":["ðŸ˜”"],"eye":["ðŸ‘€","ðŸ™„"],"grimacing":["ðŸ˜¬"],"smirk":["ðŸ˜"],"unamused":["ðŸ˜’"],"monkey":["ðŸ™ˆ","ðŸ™Š"],"kiss":["ðŸ˜˜","ðŸ’‹"],"heart":["ðŸ¥°","ðŸ˜","ðŸ˜»","ðŸ’˜","ðŸ’","ðŸ’–","ðŸ’—","ðŸ’•","ðŸ’Ÿ","â¤ï¸","ðŸ§¡","ðŸ–¤"],"angry":["ðŸ‘¿",],"smile":["ðŸ˜ˆ"],"tired":["ðŸ˜«"],"weary":["ðŸ˜©"],"disappointed":["ðŸ˜ž"],"sad":["ðŸ˜¥"],"cry":["ðŸ˜­"]}

def emoji_send(s):
  emo=False
  global nlp_text_origin
  new_text=""
  keys=["kiss","heart","angry","smile","tired","weary","disappointed","sad","cry","monkey","unamused","smirk","grimacing","eye","pensive","joy"]
  for c in s:  
    if c in emoji.UNICODE_EMOJI['en']:
      try:
        extract=demoji.findall_list(c)
        doc = nlp_text_origin(extract[0])
        for token in doc:
          # print(token,token.lemma_)
          if token.lemma_ in keys:
            emo=True
            new_text=new_text+str(token.lemma_)
            break
      except:
        pass
  return emo,new_text

import json
def get_quote():
  response = requests.get("https://zenquotes.io/api/random")
  json_data = json.loads(response.text)
  quote = json_data[0]['q'] + " -" + json_data[0]['a']
  quote="Here is a quote for you babe:\n"+quote
  return(quote)


import random
def blender_bot(message):
  if len(message.text)>1:
    emo=False
    emo,emo_key=emoji_send(message.text)
    give_text=convert_english(message.text)
    UTTERANCE = give_text
    inputs = tokenizer([UTTERANCE], return_tensors='pt')
    reply_ids = model.generate(**inputs)
    text=tokenizer.batch_decode(reply_ids)[0].replace("<s>","")
    text=text.split("</s>")[0]
    if db["output_language"]!="en":
      text=translate(text)
    if emo==True:
      your_emoji=random.choice(emoji_dict[f'{emo_key}'])
      text=text+" "+your_emoji
  else:
    text=translate(get_quote())
  return text


import re
from nltk.corpus import wordnet

def custom_talk(msg):
  msg=convert_english(msg)
  global nlp
  talk_flag=False
  list_words=['photo','timings','name']
  list_syn={}
  for your_word in list_words:
    ms = nlp.vocab.vectors.most_similar(
        np.asarray([nlp.vocab.vectors[nlp.vocab.strings[your_word]]]), n=20)
    words = {nlp.vocab.strings[w].lower() for w in ms[0][0]}
    list_syn[your_word]=words
  list_syn["selfie"]={'selfie'}
  keywords={}
  keywords_dict={}
  keywords['photo']=[]
  for synonym in list(list_syn['photo']):
      keywords['photo'].append('.*\\b'+synonym+'\\b.*')

  keywords['selfie']=[]
  for synonym in list(list_syn['selfie']):
      keywords['selfie'].append('.*\\b'+synonym+'\\b.*')

  keywords['timings']=[]
  for synonym in list(list_syn['timings']):
      keywords['timings'].append('.*\\b'+synonym+'\\b.*')
  keywords['name']=[]
  for synonym in list(list_syn['name']):
      keywords['name'].append('.*\\b'+synonym+'\\b.*')
  for intent, keys in keywords.items():
      keywords_dict[intent]=re.compile('|'.join(keys))
  responses={
    'photo':"photo",
    "selfie":"photo",
    'timings':translate(f'The time is {get_time()[0]}'),
    'name':translate(f"My name is {db['girl_name']}"),
    "fallback":"ok"}
  user_input = msg.lower()
  matched_intent = None 
  for intent,pattern in keywords_dict.items():
        if re.search(pattern, user_input): 
            matched_intent=intent  
  key='fallback' 
  if matched_intent in responses:
        key = matched_intent 
        talk_flag=False
  else:
    talk_flag=True
  return talk_flag,responses[key]  




import requests
def download_image(url):
  r = requests.get(url)
  with open(f"/content/images/{url.split('/')[-1]}", "wb") as f:
      f.write(r.content)
  return f"/content/images/{url.split('/')[-1]}"      

import random
def photos_send(bot,message):
  global gf_photos
  Gf_photo_url=random.choice(gf_photos)
  file_path=download_image(Gf_photo_url)
  supported_image=["png","jpg","jpeg"]
  extenstion=file_path.split(".")[-1]
  if extenstion.lower() in supported_image:
    bot.send_chat_action(message.chat.id, 'upload_photo')
    img = open(file_path, 'rb')
    bot.send_photo(message.chat.id, img)
    img.close()  


from gtts import gTTS 
def audio(story):
  trans_text=translate(story)
  if db["output_language"]=="en":
    tts = gTTS(text=trans_text,lang=db["output_language"],tld="ie", slow=False) 
  else:
    tts = gTTS(text=trans_text,lang=db["output_language"]) 
  save_file="/content/audio/"+story+".mp3"
  tts.save(save_file) 
  return save_file   


def format_person(text):
  text=text.lower()
  replace_word={
      "i am":"you are",
      "i was":"you were",
  }
  for i,j in replace_word.items():
    text=text.replace(i,j)
  answer_text=""
  for i in text.split():
    if i=="my":
      i=i.replace("my","your")
    if i=="mine":
      i=i.replace("mine","your")
    if i=="myself":
      i=i.replace("myself","yourself")
    if i=="me":
      i=i.replace("me","you")
    if i=="i":
      i=i.replace("i","you")
    answer_text=answer_text+i+" "
  return answer_text     
```

```python
# @title ##**main code** { display-mode: "form" }

context=""
@bot.message_handler(func=lambda m: True)
def greet(message):
  try:
    global context
    database_commad=False
    not_send=False
    check_command=message.text.lower()
    database_commad=database_check(check_command)
    # print(database_commad)
    custom_commad=False
    custom_commad=output_language_set(check_command)
    # print(custom_commad)
    if database_commad==True or custom_commad==True :
      text=f"Old message :{db['old_data']}\nlanguage : {db['output_language']}\ndatabase: {db['database']}\nn= {db['girl_name']}\ngf= {db['gf_mode']}\nmode={db['mode']}"
    else:
      if check_command[0]=="/" or check_command[0]=="?" :
        if check_command[0]=="/":
          if db["database"]==True:
            db["old_data"]=db["old_data"]+str(message.text[1:])+" "
            context=db["old_data"]
            context=convert_english(context)
            text=context
            text=translate(text)
          elif db["database"]==False:
            context=str(message.text[1:])
            context=convert_english(context)
            text=context
            text=translate(text)
        elif check_command[0]=="?":
          question = message.text[1:]
          question=convert_english(question)
          answer = qa(question=question, context=context)
          text=answer['answer']
          text=format_person(text)
          text=translate(text)
      elif check_command.startswith("mode="):
        if check_command[5:]=="c" or check_command[5:]=="v":
          if check_command[5:]=="c":
            db["mode"]="chat"
          if check_command[5:]=="v":
            db["mode"]="voice"
          text="Message mode is "+str(db["mode"])
      elif check_command.startswith("n="):
        db["girl_name"]=message.text[2:]
        text="My name is "+str(db["girl_name"])
      elif check_command.startswith("gf="):
        if check_command[3:]=="on" or check_command[3:]=="off":
          db["gf_mode"]=check_command[3:]
        text="Gf mode is "+str(db["gf_mode"])
      else:
        talk_flag=False
        talk_flag,text_back=custom_talk(check_command)
        if text_back=="photo":
          photos_send(bot,message)
          not_send=True
          text="sendig photo"
          
        else:
          text=text_back
        if talk_flag==True:
          text=blender_bot(message)
    
    if check_command=="command":
      text=f"""Change language:\nlanguage english \nlanguage bengali \nlanguage hindi \nlanguage chinese \nlanguage japanese \n [If language not change restart runtime] \n\nTemporary Database: \nenable database \ndisable database \n\nGirl name:\nn=girl_name \n\nEnable or Disable gf mode:\ngf=on\ngf=off\n\nMessage mode:\nmode=c \nmode=v"""
    if check_command=="status":
      text=f"Old message :{db['old_data']}\nlanguage : {db['output_language']}\ndatabase: {db['database']}\nn= {db['girl_name']}\ngf= {db['gf_mode']}\nmode={db['mode']}"
    if not_send==False:
      # bot.send_message(message.chat.id,text)
      if db["mode"]=="chat":
        bot.send_message(message.chat.id,text)
      elif db["mode"]=="voice":
        try:
          audio_file_path=audio(text)
          bot.send_audio(message.chat.id, audio=open(audio_file_path, 'rb'))
        except:
          bot.send_message(message.chat.id,text)
  except Exception as e:
    print(str(e))
    bot.send_message(message.chat.id,"I feel love for you")



  # print(db)
```
